```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the CSV file

```python
df=pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(['url', 'address', 'phone', 'dish_liked','reviews_list', 'menu_item'],axis=1)
df.head()
```

```python
df.info()
```

### Droping Duplicate values

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

### Removing the 'NEW', '-' and '/5' from Rate column

```python
def handleRate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
        
df['rate']=df['rate'].apply(handleRate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

### Filling NULL values in rate column with mean

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

### Droping NULL Values

```python
df.dropna(inplace=True)
df.head()
```

```python
df.info()
```

### Renaming the column

```python
df.rename(columns={'approx_cost(for two people)':'CostFor2Plates','listed_in(type)':'Type','listed_in(city)':'City'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['City'].unique()
```

```python
df=df.drop(['City'],axis=1)
```

```python
df['CostFor2Plates'].unique()
```

### Removing Comma(,) from the CostFor2Plates column

```python
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['CostFor2Plates']=df['CostFor2Plates'].apply(handlecomma)
df['CostFor2Plates'].unique()
```

```python
df.head()
```

### Cleaning Rest Type Column

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessThan1000=rest_types[rest_types<1000]
rest_types_lessThan1000
```

### Making rest_types_lessThan1000 in frequency as Other

```python
def handle_rest_type(value):
    if(value in rest_types_lessThan1000):
        return "others"
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head(10)
```

### Making location less than 300 in frequency as other

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location_lessThan300=location[location<300]

def handel_location(value):
    if(value in location_lessThan300):
        return "other"
    else:
        return value
    
df['location']=df['location'].apply(handel_location)
df['location'].value_counts()
```

```python
df.head(10)
```

### Making cuisines less than 100 in frequency as other

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines_lessThan100= cuisines[cuisines<100]

def handle_Cuisines(value):
    if value in cuisines_lessThan100:
        return "other"
    else:
        return value

df['cuisines']=df['cuisines'].apply(handle_Cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```
